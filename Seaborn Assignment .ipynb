```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
c=sns.load_dataset('flights')
```

```python
c
```

```python
import numpy as np 
```

```python
flights_pivot = c.pivot("month", "year", "passengers")
```

```python
plt.figure(figsize=(10, 8))  # Set the size of the figure

sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")  # Create the heatmap

plt.title("Number of Passengers by Year and Month")  # Set the title of the plot
plt.xlabel("Year")  # Set the label for the x-axis
plt.ylabel("Month")  # Set the label for the y-axis

plt.show()  # Show the plot
```
